```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load data
h2 = pd.read_excel("H2_data.xlsx")
o18 = pd.read_excel("O18_data.xlsx")
temp = pd.read_excel("Temperature_data.xlsx")
precip = pd.read_excel("Precipitation_data.xlsx")
```

```python
def process_df(df):
    df['Sample Date'] = df['Sample Date'].astype(str).str.strip()
    df['Sample Date'] = pd.to_datetime(df['Sample Date'], errors='coerce', utc=True)  # Convert to datetime
    
    print(df['Sample Date'].dtype)  # Debugging: Check type after conversion
    
    df['Date'] = df['Sample Date'].dt.to_period('M')  # Extract Year-Month
    df.set_index('Date', inplace=True) 
    return df

```

```python
h2 = process_df(h2)
o18 = process_df(o18)
temp = process_df(temp)
precip = process_df(precip)
```
