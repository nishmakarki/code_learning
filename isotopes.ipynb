```python
import matplotlib.pyplot as plt
import pandas as pd  
```

```python
istdeu=pd.read_excel("Deuterium.xlsx")
istoxy=pd.read_excel("oxygen_18.xlsx")
```

```python
istdeu.columns
```

```python
fig, (isoplot)=plt.subplots(1,1,figsize=(9,6))
fig.suptitle('Deuterium v/s Oxygen_18',fontsize=12)
isoplot.plot(istoxy['Measurand Amount'],istdeu['Measurand Amount'],'.',markersize=3,label='Isotopes data',color="red")
isoplot.set_xlabel('Oxygen_18',fontsize=8)
isoplot.set_ylabel('Deuterium',fontsize=8)
isoplot.set_title("Isotopes in precipitation",fontsize=10)
isoplot.legend(loc='upper right')
isoplot.grid(True)
```
