```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df=pd.read_csv("Layer_10m_depth.csv")
```

```python
df.columns
```

```python
fig, (nishma,karki)=plt.subplots(2,1,figsize=(9,7))

fig.suptitle('X and Y plot in UTM and WGS')

nishma.plot(df['X_UTM46N'],df['Y_UTM46N'],'+',markersize=12,label='Coordinate points in UTM',color="red")
nishma.set_xlabel('X_UTM64N',                     fontsize=7)
nishma.set_ylabel('Y_UTM64N')
nishma.set_title("UTM")
nishma.legend(loc='upper right')
nishma.grid(True)

karki.plot(df['X_WGS84'],df['Y_WGS84'],'+',markersize=12,label='Coordinate points in WGS', color="green")
karki.set_xlabel('X_WGS84',fontsize=7)
karki.set_ylabel('Y_WGS84')
karki.set_title("WGS")
karki.legend(loc='upper right')
karki.grid(True)

plt.tight_layout()


plt.show()
```
